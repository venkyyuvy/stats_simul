```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 30
variance = 4
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.figure(figsize=(15, 4))
plt.plot(x, stats.norm.pdf(x, mu, sigma), color='tab:orange')
plt.axvline(mu, color='tab:grey')

# sampling distribution
rng = np.random.RandomState(None)
n_samples = 2000
sample_size = 25
sample_means = np.array([np.mean(rng.normal(mu, sigma, size= sample_size)) 
                for _ in range(n_samples)])
sample_means.sort()
density =  stats.norm.pdf(sample_means, sample_means.mean(), sample_means.std(),)
plt.plot(sample_means, density ,  color='tab:blue')
plt.fill_between(sample_means, density,  color='blue', alpha=.2)
# type I error

# lb = samples.mean() + stats.norm.ppf(0.005) * (samples.std())
# plt.axvline(lb, color='violet')
# ub = samples.mean() - stats.norm.ppf(0.005) * (samples.std())
# plt.axvline(ub, color='violet')

# # null hypothesis

mu_0 = 31.5
plt.axvline(mu_0, color='tab:red')
x = np.linspace(mu_0 - 3*sigma/np.sqrt(sample_size), 
                mu_0 + 3*sigma/np.sqrt(sample_size), 100)
density = stats.norm.pdf(x, mu_0, sigma/np.sqrt(sample_size))
plt.fill_between(x, density, color='tab:red', alpha=0.2)

# # type II error
# lb = mu_0 + stats.norm.ppf(0.005) * (samples.std())
# plt.axvline(lb, color='red')
# ub = mu_0 - stats.norm.ppf(0.005) * (samples.std())
# plt.axvline(ub, color='red')

```
