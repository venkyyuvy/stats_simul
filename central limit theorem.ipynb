```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
distr = 'poisson'
size = 50000

rng = np.random.RandomState(1)
```

```python
sampler = getattr(rng, distr)
samples = sampler(3, size)
```

```python
plt.figure(figsize=(20,5))
plt.hist(samples, bins=30,  )
```

```python
sample_size = 35

n_samples = 100000

sample_sums = [rng.rand(sample_size).mean() for _ in range(n_samples)]
```

```python
plt.hist(rng.rand(sample_size))
```

```python
plt.figure(figsize=(20,5))
plt.hist(sample_sums, bins=30,)
```
