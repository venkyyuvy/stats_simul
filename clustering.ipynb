```python
from sklearn.datasets import make_circles, make_blobs, make_classification
```

```python
X , y = make_classification(n_samples=10000, n_classes=4, n_clusters_per_class=1,
                            n_informative=2, n_redundant=0, n_repeated=0, n_features=2,
                            random_state=2, class_sep=2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize': (15, 5)})
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.3)
```

```python
def plot_cluster_output(X, clusterer, ax):
    ax.scatter(X[:, 0], X[:, 1], c=clusterer.labels_)
    if isinstance(clusterer, KMeans):
        centroids = clusterer.cluster_centers_
        ax.scatter(centroids[:, 0], centroids[:, 1], c='black', s=10)
```

```python
from sklearn.cluster import KMeans
```

```python

fig, axes = plt.subplots(1, 7)
for iter, ax in enumerate(axes):
    print(f'{iter=}')
    kmeans = KMeans(n_init=1, init='random',
        max_iter=iter+1, n_clusters=4, random_state=21).fit(X)
    print('score:', kmeans.inertia_)
    plot_cluster_output(X, kmeans, ax)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python

fig, axes = plt.subplots(1, 4)
for ax, method in zip(axes, ['ward', 'single', 'complete', 'average']):
    aggl = AgglomerativeClustering(n_clusters=4, linkage=method).fit(X)
    plot_cluster_output(X, aggl, ax)
    ax.set_title(method, fontsize=20)
```
